```python
def add_to_balance_sheet1(table):
   
   d = pd.DataFrame()
   
   k1 = table[table.index.str.contains(r"Total.*assets",case=False, regex=True,na=False)]

   if len(k1) > 0:
      d["Total assets"] = k1.iloc[0,]

   k1 = table[table.index.str.contains(r"Total.*liabilities",case=False, regex=True, na=False)]

   if len(k1) > 0:
      d["Total liabilities"] = k1.iloc[0,]
    
   k1 = table[table.index.str.contains(r"Total.*equity",case=False, regex=True, na=False)]

   if len(k1) > 0:
      d["Total shareholder equity"] = k1.iloc[0,] 

   k1 = table[table.index.str.contains(r"Total liabilities and.*equity",case=False, regex=True, na=False)]

   if len(k1) > 0:
      d["Total liabilities and stockholders equity"] = k1.iloc[0,] 
   
   return d
```

```python
def clean_table (table):
    
    df = table.copy()
    
    if df[df.columns[1]].dtype == object :
       df[df[df.columns[1]].str.contains((r".*[\d]"), case=False, regex=True,na=False)] = np.nan
    col = list(df.columns)
    col[0] = ''
    for i in range(1,len(col)):
      col[i] = col[i][-4:]
    df.columns = col  
    df.set_index(df.columns[0], inplace=True)
    df.dropna(axis=1, how='all', inplace=True)
    return df
```

```python
rootdir = "C:/Users/priye/Desktop/Dissertation/DATA/"

bal_list = []

for d in listdir(rootdir):
  balance_sheet = pd.DataFrame()
  dirname = rootdir + d
  for f in listdir(dirname):
    if f.endswith('xlsx'):
      print(f)
      path = dirname + '/' + f
    
      xl_file = pd.ExcelFile(path)
      table = xl_file.parse(company.loc[f,'Balance'])
      ct = clean_table(table)
      df = add_to_balance_sheet1(ct) 
      balance_sheet = balance_sheet.append(df)
  balance_sheet = balance_sheet.T
  balance_sheet['company'] = d
  bal_list.append(balance_sheet)
 
```
