```python
import numpy as np
import pandas as pd
import openpyexcel
from openpyexcel import workbook, load_workbook
from openpyexcel.utils import get_column_letter
from itertools import islice
```

```python
from os import listdir
import re
```

```python
def alloc_sheet(f,slist,company):

  '''
  Identify income, balance and cash sheet from list of sheets provided (slist).
  '''

  # Balance

  r = re.compile("Consolidated[\s_]Balance[\s_]Sheet[s]?$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]
   
  r = re.compile("Consolidated[\s_]Financial[\s_](Positio|Position)[s]?$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))
    
  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]
   
  r = re.compile("Consolidated[\s_]Statement[s]?[\s_]of[\s_](Fi|Fin|Fina)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]
    
  r = re.compile("Consolidated Balance Sheets, as$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]
    
  r = re.compile("Consolidated Balance Sheets - S$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]

  r = re.compile("Consolidated[\s_]Balance[\s_]Sheets[\s_]\(Fi$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Balance'] = filtered_list[0]

  # Income
  
  r = re.compile("Consolidated[\s_]Statement[s]?[\s_]of[\s_](Ea|Ear|Earn|In|Inc|Inco|Incom|Income|Ope|Oper|Opera)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))
    
  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]

  #r = re.compile("Consolidated[\s_]Statement[s]?[\s_]of[\s_](Com|Comp|Compr|Compre)$",re.IGNORECASE)
  #filtered_list = list(filter(r.match, slist))
    
  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]

  r = re.compile("Consolidated[\s_]Results[\s_]of[\s_](Ope|Oper|Opera|Operat|Operati)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]
    
  r = re.compile("Consolidated[\s_]Income.*Statement[s]?$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))
   
  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]
    
  r = re.compile("CONSOLIDATED[\s_]AND[\s_]SECTOR[\s_]INCOME$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))
    
  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]

  r = re.compile("Statements[\s_]of[\s_]Consolidated[\s_](Inc|Inco|Incom|Income|Ear|Earn)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Income'] = filtered_list[0]
    
  # Cash
    
  r = re.compile("Consolidated[\s_]Statement[s]?[\s_]of[\s_](Cash\s|Cash|Cas|Ca)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Cash'] = filtered_list[0]
    
  r = re.compile("Consolidated[\s_]Statement[s]?[\s_]of[\s_]Cash flows \(USD \$\)$",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Cash'] = filtered_list[0]
    
  r = re.compile("Statements[\s_]of[\s_]Consolidated[\s_](Ca|Cas|Cash$)",re.IGNORECASE)
  filtered_list = list(filter(r.match, slist))

  if len(filtered_list) > 0:
    company.loc[f,'Cash'] = filtered_list[0]
```

```python
rootdir = "C:/Users/priye/Desktop/Vault/Projects/TS1/DATA/"

sheet={'a','b'}
company = pd.DataFrame()

for d in listdir(rootdir):
    dirname = rootdir + d
    for f in listdir(dirname):
        if f.endswith('xlsx'):
             print(f)
             path = dirname + '/' + f
             wb = load_workbook(path)
             slist = wb.sheetnames
             alloc_sheet(f,slist,company)
             wb.close()
```
